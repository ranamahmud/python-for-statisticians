# setting style

```python
# sns.set_context("notebook") 
sns.set_style('darkgrid')
sns.distplot(df['mpg'])
plt.show()
```

```python
sns.set_style('whitegrid')
sns.distplot(df['mpg'])
# sns.despine(left=True)
plt.show()
```

```python
# removing splines
sns.set_style('whitegrid')
sns.distplot(df['mpg'])
sns.despine(left=True, right = True)
plt.show()
```
